```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
d =  pd.read_csv("bitcoin_difficulty.csv")
```

```python
data = pd.read_csv("bitcoin_difficulty.csv",usecols=['Time','Mining Difficulty'], parse_dates=['Time'])
price = pd.read_csv("BTC-USD.csv",usecols=["Date","Price"], parse_dates=["Date"])
```

```python
data.set_index('Time', inplace = True)
price.set_index("Date", inplace = True)
```

```python

data1 = data['Mining Difficulty']
data2 = price["Price"]
fig, ax1 = plt.subplots(figsize=(15,7))

color = 'firebrick'

ax1.set_ylabel('Difficulty in Trillion(tera)', color=color, fontsize = 15)
ax1.set_xlabel("Year",fontsize = 15,alpha = .9)
ax1.plot( data1, color=color, linewidth = 3,alpha = .7)
ax1.set_yticklabels([ '0','1', '1T', '2T', '3T', '4T', '5T', '6T',' 7T'])

ax1.tick_params(axis='y', labelcolor=color)



ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'darkorange'
ax2.set_ylabel('Price in US-dollar', color=color, fontsize = 15, )  # we already handled the x-label with ax1
ax2.plot( data2, color=color, linewidth = 3,alpha = .7)
ax2.tick_params(axis='y', labelcolor=color)

ax1.set_ylabel('Diffuculty in Trillion(Tera)')
ax2.set_ylabel('Price in US Dollar')
lines = ax1.get_lines() + ax2.get_lines()
ax1.legend(lines, [l.get_label() for l in lines], loc='upper left', fontsize = 15)


plt.title("Bitcoin Mining Difficulty and Price from 2009",fontsize = 20)
plt.text("2017-06-01", 450, "Source: data.bitcoinity.org,\n            finance.yahoo.com",fontsize = 12)

plt.style.use('default')
fig.tight_layout()  # ght y-label is slightly clipped
plt.annotate("",xy = ("2018-11-30", 6600), xytext = ("2018-11-30", 13500), arrowprops = dict(facecolor = 'orangered',width = 4,headwidth =20, alpha = .8
                                                                                    
                                                                                            ))
plt.annotate("", xy = ("2018-11-30", 13900),xytext = ("2018-11-30", 7000), arrowprops = dict(facecolor = 'orangered',width = 4,headwidth=20,alpha = .8))
plt.annotate("This arrow will be bigger?",fontsize = 17, xy = ("2018-10-30", 10000),xytext = ("2014-03-30", 7500), arrowprops = dict(connectionstyle="angle3", facecolor = 'black',alpha = .40,width = 2,headwidth=13))
plt.savefig('BTC_price_miningdifficulty.jpg')
plt.show()
```
